```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from datetime import *

# load data and extract relevant features
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
traffic_data = pd.read_csv("traffic.csv")
vancouver_weather = pd.read_csv("vancouver.csv")
victoria_weather = pd.read_csv("victoria.csv")[['Day', 'Date.Time', 'Month', 'Time', 'Wind.Speed.km.per.h', 'Year']].rename(columns={'Date.Time':'Full.Date'})

tsawassen_weather = pd.read_csv("tsa_weather.csv")[['Day', 'Full.Date', 'Month', 'Time', 'Wind.Speed.km.per.h', 'Year']]
horseshoebay_weather = pd.read_csv("pa_weather.csv")[['Day', 'Full.Date', 'Month', 'Time', 'Wind.Speed.km.per.h', 'Year']]
nanaimo_weather = pd.read_csv("nan_weather.csv")[['Day', 'Full.Date', 'Month', 'Time', 'Wind.Speed.km.per.h', 'Year']]



def reformat_time(data):
    data['Time'] = data['Time'].str.replace(':', '').astype(int)
    data['Time.Hour'] = data['Time'].astype(str).str[:-2]
    return data['Time.Hour']

# extract day from vancouver weather files
def extract_day(time):
    return datetime.strftime(datetime.strptime(time, '%Y-%m-%d %H:%M:%S'), "%d")

# extract day from other weather files
def extract_day2(time):
    return datetime.strftime(datetime.strptime(time, '%Y-%m-%d %H:%M'), "%d")

victoria_weather['Day.of.Month'] = victoria_weather['Full.Date'].apply(extract_day).astype(int)
tsawassen_weather['Day.of.Month'] = tsawassen_weather['Full.Date'].apply(extract_day2).astype(int)
nanaimo_weather['Day.of.Month'] = nanaimo_weather['Full.Date'].apply(extract_day2).astype(int)
horseshoebay_weather['Day.of.Month'] = horseshoebay_weather['Full.Date'].apply(extract_day2).astype(int)

victoria_weather['Region.Code'] = 4
tsawassen_weather['Region.Code'] = 3
nanaimo_weather['Region.Code'] = 2
horseshoebay_weather['Region.Code'] = 1

weather_data = pd.concat([victoria_weather, tsawassen_weather, nanaimo_weather, horseshoebay_weather])
weather_data['Time.Hour'] = reformat_time(weather_data)

weather_data.to_csv('processed_weather_data.csv', index=False)
```
